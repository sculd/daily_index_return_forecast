```python
from datetime import datetime, timedelta

import yfinance as yf
import pandas as pd
import numpy as np
```

```python
start_date = "2023-01-01"
end_date = (datetime.today() + timedelta(days = 1)).strftime("%Y-%m-%d")
```

```python
df_spy = yf.download("SPY", start=start_date, end=end_date)
df_xlc = yf.download("XLC", start=start_date, end=end_date)
df_xly = yf.download("XLY", start=start_date, end=end_date)
df_xlp = yf.download("XLP", start=start_date, end=end_date) 
df_xle = yf.download("XLE", start=start_date, end=end_date) 
df_xlf = yf.download("XLF", start=start_date, end=end_date)
df_xlv = yf.download("XLV", start=start_date, end=end_date) 
df_xli = yf.download("XLI", start=start_date, end=end_date)
df_xlb = yf.download("XLB", start=start_date, end=end_date)
df_xlre = yf.download("XLRE", start=start_date, end=end_date)
df_xlk = yf.download("XLK", start=start_date, end=end_date) 
df_xlu = yf.download("XLU", start=start_date, end=end_date)
```

```python
dfs = [df_spy, df_xlc, df_xly, df_xlp, df_xle, df_xlf, df_xlv, df_xli, df_xlb, df_xlre, df_xlk, df_xlu]

for df in dfs:
    df["Adjustment Multiplier"] = df["Adj Close"] / df["Close"]
    df["Adj Open"] = df["Open"] * df["Adjustment Multiplier"]
    df["returns"] = ((df["Adj Open"] - df["Adj Close"].shift(1)) / df["Adj Close"].shift(1)).fillna(0)
```

```python
df_returns = pd.concat(
    [df_xlc.add_prefix("Communications_"), df_xly.add_prefix("ConsumerDiscretionary_"), df_xlp.add_prefix("ConsumerStaples_"), 
     df_xle.add_prefix("Energy_"), df_xlf.add_prefix("Financials_"), df_xlv.add_prefix("Healthcare_"), df_xli.add_prefix("Industrials_"), 
     df_xlb.add_prefix("Materials_"), df_xlre.add_prefix("RealEstate_"),df_xlk.add_prefix("Technology_"), df_xlu.add_prefix("Utilities_"), 
     df_spy.add_prefix("SP500_")], axis = 1).dropna()

df_returns = df_returns[["Communications_returns","ConsumerDiscretionary_returns", "ConsumerStaples_returns", "Energy_returns",
                         "Financials_returns", "Healthcare_returns", "Industrials_returns",
                         "Materials_returns","RealEstate_returns", "Technology_returns",
                         "Utilities_returns","SP500_returns"]]
df_returns = df_returns.iloc[1:]
df_returns["SP500_forward_returns"] = df_returns["SP500_returns"].shift(-1)
df_returns = df_returns.dropna()
```
